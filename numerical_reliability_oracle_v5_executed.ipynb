4. Algebraic surgery: localizes a matmul fault to output block and reduction band.
5. SPRT evidence ledger with alpha=1e-5, beta=0.02 for gray-zone structured drift.
6. Co-evolutionary arena: mutators are optimized to evade the oracle, then measured.
